```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns
```

```python
df = pd.read_csv("worldbank-country-internet-data.csv")
df.head()
```

```python
df.columns
```

```python
df.columns = [col[:4] if '[' in col else col for col in df.columns ]
```

```python
df.columns
```

```python
selected_columns = ["Country Name","Country Code","Series Name","Series Code"]
selected_years = [str(year) for year in range(1990,2024)]

selected_years
```

```python
df = df[selected_columns+selected_years]
```

```python
df.head()
```

```python
df.to_csv("cleaned_world_bank.csv",index=False)
```
