```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df
```

```python
temp = sorted(df["sepal_length"].tolist())
temp
```

```python
y_quant = []
for i in range(0,101):
    y_quant.append(np.percentile(temp,i))
```

```python
samples = np.random.normal(loc=0,scale=1,size=100)
```

```python
x_quant=[]
for i in range(0,101):
    x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

### Other way for plotting 

```python
import statsmodels.api as sm

# create a QQplot of the two sets of data
fig = sm.qqplot(df["sepal_length"],line="45",fit=True)


# Adda title and labels to the plot
plt.title("QQplot")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Actaul quantiles")

plt.show()
```

### QQplot can be used for comparing any distribution , generally used for normal distribution
